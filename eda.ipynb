```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the dataset
file_path = '/content/drive/MyDrive/ENPM703/original/Dataset/Metadata/Metadata/Metadata_Participants.csv'
metadata = pd.read_csv(file_path)

# 1. Display the first few rows of the dataset for inspection
print("First few rows of the dataset:")
print(metadata.head())

# 2. Check for missing values
print("\nChecking for missing values:")
missing_values = metadata.isnull().sum()
print(missing_values)

# 3. Summary statistics
print("\nSummary statistics:")
summary_stats = metadata.describe()
print(summary_stats)

# 4. Visualize age distribution by class
plt.figure(figsize=(8, 6))
sns.histplot(data=metadata, x='Age', hue='Class', kde=True, bins=10, alpha=0.6)
plt.title('Age Distribution by Class')
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.show()

# 5. Gender distribution
plt.figure(figsize=(6, 6))
gender_counts = metadata['Gender'].value_counts()
gender_counts.plot(kind='pie', autopct='%1.1f%%', labels=['Male', 'Female'], startangle=140)
plt.title('Gender Distribution')
plt.ylabel('')
plt.show()

# 6. CARS Score distribution for ASD group
plt.figure(figsize=(8, 6))
asd_group = metadata[metadata['Class'] == 'ASD']
sns.histplot(asd_group['CARS Score'], kde=True, bins=10, color='blue', alpha=0.6)
plt.title('CARS Score Distribution (ASD Group)')
plt.xlabel('CARS Score')
plt.ylabel('Frequency')
plt.show()

# 7. Boxplot for Age vs Class
plt.figure(figsize=(8, 6))
sns.boxplot(data=metadata, x='Class', y='Age', palette="Set2")
plt.title('Age Distribution by Class')
plt.xlabel('Class')
plt.ylabel('Age')
plt.show()

# 8. Correlation Heatmap (Numerical Data)
plt.figure(figsize=(8, 6))
correlation_matrix = metadata[['Age', 'CARS Score']].corr()
sns.heatmap(correlation_matrix, annot=True, cmap="coolwarm", fmt=".2f")
plt.title('Correlation Heatmap')
plt.show()

print("\nEDA Completed Successfully!")


plt.figure(figsize=(8, 6))
tc_group = metadata[metadata['Class'] == 'TS']
sns.histplot(tc_group['Age'], kde=True, bins=10, color='green', alpha=0.6)
plt.title('Age Distribution (TS Group)')
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.show()

# TS (Autism Spectrum Disorder) Age Distribution
plt.figure(figsize=(8, 6))
ts_group = metadata[metadata['Class'] == 'TC']
sns.histplot(ts_group['Age'], kde=True, bins=10, color='orange', alpha=0.6)
plt.title('Age Distribution (ASD Group)')
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.show()


plt.figure(figsize=(8, 6))
sns.histplot(tc_group['Age'], kde=False, bins=10, color='green', alpha=0.6)
plt.title('Age Distribution (TS Group)')
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.show()

plt.figure(figsize=(8, 6))
sns.histplot(ts_group['Age'], kde=False, bins=10, color='orange', alpha=0.6)
plt.title('Age Distribution (ASD Group)')
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.show()
```
