```python
import cv2
import numpy as np
import os
from sklearn.model_selection import train_test_split
import shutil
```

```python
def preprocess_image(image_path):
    img = cv2.imread(image_path)
    img_resized = cv2.resize(img, (100, 100))
    img_gray = cv2.cvtColor(img_resized, cv2.COLOR_BGR2GRAY)
    img_normalized = img_gray / 255.0
    binary_mask = (img_normalized > 0).astype(np.uint8)
    binary_mask = binary_mask * 255
    return binary_mask
```

```python
image_folder = '/content/drive/MyDrive/ENPM703/Project/Final/Dataset/TC/Raw/'
processed_folder = '/content/drive/MyDrive/ENPM703/Project/Final/Dataset4/TC/'
image_files = [f for f in os.listdir(image_folder) if os.path.isfile(os.path.join(image_folder, f))]
if not os.path.exists(processed_folder):
    os.makedirs(processed_folder)
for f in image_files:
    image_path = os.path.join(image_folder, f)
    processed_image = preprocess_image(image_path)
    processed_path = os.path.join(processed_folder, f)
    cv2.imwrite(processed_path, processed_image)
```
